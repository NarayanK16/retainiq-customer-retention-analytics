## let's do cleaning process


```python
df_clean = df.copy()
```

```python
df_clean = df_clean.drop(['customerID'],axis=1)
```

```python
df_clean.head()
```

## converting the data types of columns

```python
df_clean['TotalCharges'] = pd.to_numeric(df_clean['TotalCharges'],errors='coerce')
df_clean['MonthlyCharges'] = pd.to_numeric(df_clean['MonthlyCharges'],errors='coerce')
```

```python
df_clean['MonthlyCharges'].dtype
```

### checking null values in 'TotalCharges' column and filling the null values with 0

```python
df_clean['TotalCharges'].isnull().sum()
```

```python
df_clean['TotalCharges'].fillna(0,inplace=True)
```

```python
df_clean['TotalCharges'].isna().sum()
```

```python
df_clean['MonthlyCharges'].isna().sum()
```

## mapping the column values to make it more clean

```python
#df_clean.columns
replace_cols = ['OnlineSecurity','OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
       'StreamingMovies','MultipleLines']
for col in replace_cols:
    df_clean[col] = df_clean[col].replace({"No internet service":"No",'No phone service':"No"})
```

```python
df_clean['Churn'] = df_clean['Churn'].map({"Yes":1,"No":0})
```
